```python
#year end frequency
df.resample(rule='A').mean()
```

```python
df['High'].resample(rule='A').mean().plot(figsize=(10,5),color='red',grid=True,marker='o',title='Year end Trends Analysis',legend='High')
```

```python
#Quarterly Frequency
df.resample(rule='QS').mean()
```

```python
df['High'].resample(rule='QS').mean().plot(figsize=(12,4),title='Quarterly wise Trend Analysis',color='black',legend='High')
```

```python
#Business Year end Frequency
df.resample(rule='BA').max()
```

```python
#Business year end highest price
df['Low'].resample(rule='BA').max().plot(figsize=(10,5),title='Business year end Maximum of low share',color='green')
```

```python
#Business day Quarterly wise share details
df.resample(rule='BQS').mean()
```

```python
df['Close'].resample(rule='BQS').mean().plot(figsize=(12,5),title='Average Quarterly Close Shares',color='orange',legend='Close')
```

```python
#Monthly wise shares
df.resample(rule='M').mean()
```

```python
df['Open'].resample(rule='M').mean().plot(kind='bar',figsize=(20,7),title='Monthly wise open shares',grid=True,color='violet',fontsize=15)
```

```python
#Rolling window calculations
#Simple Moving Average
df.rolling(window=10).mean().head(10)
```

```python
df['High'].rolling(window=10).mean().plot()
```

```python
df[['Open','High']].rolling(window=10).mean().plot(color=['blue','yellow'],figsize=(12,4),title='Moving average on High and Open shares')
```

```python
df_m=df
```

```python
#Comparing the Moving Average with actual growth
df_m['Low_10days_rolling_average']=df_m['Low'].rolling(10).mean()
```

```python
df_m[['Low','Low_10days_rolling_average']].plot(figsize=(9,5),xlim=['2020-01-01','2020-12-31'],title='Trend growth on Low shares and Moving Average on share ')
```

```python
#Cummulative Moving Average
#CMA
#Expanding Function
```

```python
df_m['CMA_Open']=df_m['Open'].expanding().mean()
```

```python
#Cummulative average on open share vs actual share data
df_m[['Open','CMA_Open']].plot(figsize=(9,5),title='Cummulative average on open share vs actual share data')
```

```python
#Exponential Moving Average
#EMA
#function ewm
df_m['EMA_Close']=df_m['Close'].ewm(alpha=0.1,adjust=False).mean()
```

```python
df_m[['Close','EMA_Close']].plot(figsize=(9,5),title='Exponential Moving Average on Close share vs actual closed share')
```
