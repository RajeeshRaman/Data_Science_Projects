# bar plot for top 10 country

```python
plt.rcParams['figure.figsize']=(10,5)
plt.bar(list(fifa['nationality'].value_counts().keys()[0:10]),list(fifa['nationality'].value_counts()[0:10]),color=['green','purple'],alpha=0.2)
plt.title('Number of Players in each Country')
plt.xlabel('Country')
plt.ylabel('Players Count')
plt.grid()
plt.show()
```

# sorting the Player_salary based on highest salary

```python
# no need value_counts for player_salary since we are taking only 2 columns
```

```python

plt.figure(figsize=(21,6))
plt.bar(list(player_salary['short_name'].value_counts().keys()[0:10]),list(player_salary['wage_eur'].value_counts()[0:10]),color='gray',alpha=1)
plt.xlabel('players',fontsize=20,color='y')
plt.ylabel('salary',fontsize=20,color='y')
plt.title(label='Top 10 paid players',fontsize=40,color='g')

plt.show()
```
