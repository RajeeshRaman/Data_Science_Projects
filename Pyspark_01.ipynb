```python
df_spark=spark.read.csv('/content/drive/MyDrive/Pyspark/test2.csv',header=True,inferSchema=True)
```

```python
df_spark.show()
```

```python
# Dropping null values
```

```python
#drop all null rows in an data set
df_spark.na.drop().show()
```

```python
#droping on particular column
# how
# 'any' delete any row even single null value is present
# 'all' delete the row with all the column have null values
df_spark.na.drop(how='any').show()
```

```python
# thresh
# atleast 2 non null value should be there else it delete the row
df_spark.na.drop(how='any',thresh=2).show()
```

```python
# subset
# we can specify an specific column to remove null values
df_spark.na.drop(how='any',subset=['age']).show()
```

```python
# fill the missing values
```

```python
df_spark.na.fill("Missing Values").show()
```

```python
df_spark.na.fill('miss').show()
```

```python
df_spark.na.fill(0,subset='Experience').show()
```

```python
# replacing the missing value with mean,median,mode
```

```python
# Using Imputer function
```

```python
from pyspark.ml.feature import Imputer
```

```python
imputer=Imputer(
  inputCols=['age','Experience','Salary'],
  outputCols=["{}_imputed".format(c) for c in ['age','Experience','Salary']]

).setStrategy('mean') # change the setStrategy to mean,median,mode
```

```python
# Add Imputed Col to df
imputer.fit(df_spark).transform(df_spark).show()
```

Pyspark in ML

```python
machine=spark.read.csv('/content/drive/MyDrive/Pyspark/test1.csv',header=True,inferSchema=True)
```

```python
machine.show()
```

```python
!pip install Mllib
```

```python
machine.columns
```

```python
from pyspark.ml.feature import VectorAsasembler
```

```python
featureassembeler=VectorAssembler(inputCols=['age', 'Experience'],outputCol='Independent Feature')
```

```python
output=featureassembeler.transform(machine)
```

```python
output.show()
```

```python
output.columns
```

```python
final_data=output.select(['Salary','Independent Feature'])
```

```python
final_data.show()
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
#train Test Split
train_data,test_data=final_data.randomSplit([0.75,0.25])
```

```python
regressor=LinearRegression(featuresCol='Independent Feature',labelCol='Salary')
```

```python
#fitting the train data
regressor=regressor.fit(train_data)
```

```python
# Coefficient
regressor.coefficients
```

```python
# intercepts
regressor.intercept
```

```python
# Prediction
pred_result=regressor.evaluate(test_data)
```

```python
pred_result.predictions.show()
```

```python
pred_result.meanAbsoluteError,pred_result.meanSquaredError
```
